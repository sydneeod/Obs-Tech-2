What exactly is a grating?

Generally, gratings are finely etched surfaces.  CDs and DVDs - the now-obsolete storage media - are finely grooved and behave as gratings.



A reflection grating works the same way.

A reflection grating can be etched with a *blaze angle* such that most of the light ends up in a specific order *other than* zero'th order (which is non-dispersive and can't be used for spectroscopy)

# Volume Phase Holography Grating

VPH gratings, like those Bo Zhao showed, follow the same equations as regular transmission gratings, except they put all of their light into a single dispersive order.

In other words, no light gets "lost" to zero-order.

See, e.g., https://www.noao.edu/noao/noaonews/jun98/node4.html for more details.

# What does spectroscopic data look like?

The spectrograph takes light and spreads it out to different positions on the detector.

A two-dimensional slit spectrum will look like this:

# A 2D Spectrum

Wavelength is along the X-axis: this is the "dispersion axis"

Position is along the Y-axis.  In the case of a single star, position isn't very important!  You're just looking at different parts of the *point spread function* (PSF).

# A 2D Spectrum
